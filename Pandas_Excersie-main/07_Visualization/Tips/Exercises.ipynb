### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv("tips.csv")
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
tips = tips.drop(columns="Unnamed: 0")
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
ax = tips["total_bill"].plot(kind="hist", bins=20, figsize=(7, 4), color="steelblue")
ax.set_xlabel("Total bill")
ax.set_title("Distribution of Total Bill")
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
ax = sns.scatterplot(data=tips, x="total_bill", y="tip")
ax.set_title("Total Bill vs Tip")
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.pairplot(tips[["total_bill", "tip", "size"]])
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
ax = sns.stripplot(data=tips, x="day", y="total_bill", jitter=True)
ax.set_title("Total Bill by Day")
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
ax = sns.stripplot(data=tips, x="tip", y="day", hue="sex", jitter=True)
ax.set_title("Tip by Day and Sex")
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
ax = sns.boxplot(data=tips, x="day", y="total_bill", hue="time")
ax.set_title("Total Bill by Day and Time")
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
g = sns.FacetGrid(tips, col="time", height=4, aspect=1.1)
g.map_dataframe(sns.histplot, x="tip", bins=10)
g.set_axis_labels("Tip", "Count")
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    col="sex",
    kind="scatter",
)
plt.show()
```

### BONUS: Create your own question and answer it using a graph.

```python
# Bonus: Which day has the highest average tip percentage?
tips["tip_pct"] = tips["tip"] / tips["total_bill"]
ax = sns.barplot(data=tips, x="day", y="tip_pct")
ax.set_title("Average Tip Percentage by Day")
ax.set_ylabel("Tip percentage")
plt.show()
```
