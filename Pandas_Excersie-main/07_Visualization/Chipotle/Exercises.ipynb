### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv("chipotle.tsv", sep="\t")
chipo["item_price"] = chipo["item_price"].str.replace("$", "", regex=False).astype(float)
chipo.head()
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
top_items = chipo.groupby("item_name")["quantity"].sum().sort_values(ascending=False).head(5)

ax = top_items.plot(kind="bar", figsize=(8, 4), color="steelblue")
ax.set_xlabel("Item")
ax.set_ylabel("Quantity ordered")
ax.set_title("Top 5 Chipotle Items Ordered")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
orders = (
    chipo.assign(revenue=chipo["quantity"] * chipo["item_price"])
    .groupby("order_id")
    .agg(items=("quantity", "sum"), order_price=("revenue", "sum"))
)

ax = orders.plot.scatter(x="order_price", y="items", figsize=(8, 5), alpha=0.5)
ax.set_xlabel("Order price")
ax.set_ylabel("Items ordered")
ax.set_title("Items Ordered per Order Price")
plt.show()
```

### Step 7. BONUS: Create a question and a graph to answer your own question.

```python
# Bonus: Which items generated the most revenue?
item_revenue = (
    chipo.assign(revenue=chipo["quantity"] * chipo["item_price"])
    .groupby("item_name")["revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

ax = item_revenue.plot(kind="barh", figsize=(8, 5), color="seagreen")
ax.set_xlabel("Revenue")
ax.set_ylabel("Item")
ax.set_title("Top 10 Items by Revenue")
plt.tight_layout()
plt.show()
```
