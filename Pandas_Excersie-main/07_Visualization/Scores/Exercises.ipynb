### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### Step 2. Create the DataFrame that should look like the one below.

```python
raw_data = {
    "first_name": ["Jason", "Molly", "Tina", "Jake", "Amy"],
    "last_name": ["Miller", "Jacobson", "Ali", "Milner", "Cooze"],
    "sex": ["male", "female", "male", "female", "female"],
    "preTestScore": [4, 24, 31, 2, 3],
    "postTestScore": [25, 94, 57, 62, 70],
    "age": [42, 52, 36, 24, 73],
}

df = pd.DataFrame(
    raw_data,
    columns=["first_name", "last_name", "sex", "preTestScore", "postTestScore", "age"],
)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
ax = df.plot.scatter(
    x="preTestScore",
    y="postTestScore",
    s=df["age"],
    figsize=(7, 5),
    alpha=0.7,
)
ax.set_xlabel("Pre-test Score")
ax.set_ylabel("Post-test Score")
ax.set_title("Pre-test vs Post-test Score by Age")
plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
colors = df["sex"].map({"male": "steelblue", "female": "tomato"})

ax = df.plot.scatter(
    x="preTestScore",
    y="postTestScore",
    s=df["postTestScore"] * 4.5,
    c=colors,
    figsize=(7, 5),
    alpha=0.7,
)
ax.set_xlabel("Pre-test Score")
ax.set_ylabel("Post-test Score")
ax.set_title("Pre-test vs Post-test Score by Sex")
plt.show()
```
